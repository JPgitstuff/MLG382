```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Load the dataset
dataset = pd.read_csv('AllProductReviews.csv')

# Ensure the dataset has a column named 'text' for analysis
if 'ReviewBody' in dataset.columns:
    # Perform sentiment analysis on each row
    analyzer = SentimentIntensityAnalyzer()
    dataset['sentiment_scores'] = dataset['ReviewBody'].apply(lambda x: analyzer.polarity_scores(x))
    print(dataset[['sentiment_scores']])
else:
    print("The dataset does not contain a 'text' column.")
```

```python
import numpy as np
import pandas as pd
nltk.download('movie_reviews')
from nltk.corpus import movie_reviews
```

```python
data = [(list(movie_reviews.words(fileid)), category) 
for category in movie_reviews.categories() 
for fileid in movie_reviews.fileids(category)]
```

```python
#Print the first 100 records 
for i, (words, category) in enumerate(data[:100]):
         print(f"Record {i+1}:")    
         print(f"Category: {category}")     
         print(f"Words: {words[:20]} ...")  # Print only the first 20 words for brevity     print("-" * 50)
```

```python
# Save the results to a new CSV file
data_df = pd.DataFrame(data, columns=['words', 'category'])
data_df.to_csv('movie_reviews.csv', index=False)

```
