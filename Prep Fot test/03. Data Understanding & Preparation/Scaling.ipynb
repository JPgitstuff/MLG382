```python
import pandas as pd
from sklearn.preprocessing import RobustScaler
import seaborn as sns 
import matplotlib.pyplot as plt

# Sample data with an outlier
data = {'Feature': [10, 12, 14, 16, 18, 15,1, 2, 2, 3, 3, 4, 5, 10, 15, 25,25, 20, 18, 15, 12, 10, 8, 7, 6, 5, 5,1000]}  # 1000 is an outlier
df = pd.DataFrame(data)


```

```python
sns.boxplot(data=df, x='Feature')
```

```python
sns.histplot(data=df, x='Feature')
```

```python
# Apply RobustScaler
scaler = RobustScaler()
scaled_data = scaler.fit_transform(df)

# Show results
scaled_df = pd.DataFrame(scaled_data, columns=['Feature_Scaled'])
print(pd.concat([df, scaled_df], axis=1))
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=scaled_df)
plt.show()
```
