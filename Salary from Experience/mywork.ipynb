```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
filename  =  'SalariesVsExperience.xlsx'
df = pd.read_excel(filename)
df # or print(df) deos the same
```

```python
df.plot.scatter(x = 'Years Experience', y = 'Salary')
## sns.regplot(data=df, x='Years Experience', y='Salary') adds the regression line import seaborn
```

```python
df.corr()
```

```python
from sklearn.linear_model import LinearRegression
df['Years Experience'] # gives all the values in that column
```

```python
X = df['Years Experience'].to_frame() #Converts into a dataframe variable
Y = df['Salary'].to_frame()
```

```python
model = LinearRegression() # instansiates the modle 
model.fit(X,Y) #fit data to the modle 
## machine learing is done here
```

```python
xn = np.linspace(0,10,201)
x_pred = pd.DataFrame({'Years Experience' : xn})
y_pred = model.predict(x_pred)
xn
```

```python
plt.scatter(X,Y,color = 'blue', s = 10)
plt.plot(x_pred, y_pred, color = 'red')
plt.xlabel('Years Experience')
plt.ylabel('Salary (R)')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
y_p = model.predict(X) ## use X and y valuesfor plotting not the predictions
mae = mean_absolute_error(Y,y_p)
mse = mean_squared_error(Y, y_p)
rmse = np.sqrt(mse)

print(f'MAE = {mae}')
print(f'MSE = {mse}')
print(f'RMSE = {rmse}')
```
